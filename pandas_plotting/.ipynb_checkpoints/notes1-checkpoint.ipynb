```python
import pandas as pd
import numpy as np
from pandas_datareader import data

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
aapl = data.DataReader('aapl', 'google')
msft = data.DataReader('msft', 'google')
amd = data.DataReader('amd', 'google')
aapl = aapl['Close']
msft = msft['Close']
amd = amd['Close']
aapl.name = 'Apple'
msft.name = 'Microsoft'
amd.name = 'AMD'
```

```python
aapl.head()
```

```python
aapl.plot(legend=True)
msft.plot(legend=True)
amd.plot(legend=True)
```

```python
aapl0 = (aapl - aapl.iloc[0]) / aapl.iloc[0]
msft0 = (msft - msft.iloc[0]) / msft.iloc[0]
amd0 = (amd - amd.iloc[0]) / amd.iloc[0]
aapl0.name = 'Apple'
msft0.name = 'Microsoft'
amd0.name = 'AMD'
zero = aapl * 0
```

```python
zero.plot(style='--k')
aapl0.plot(legend=True, figsize=(12, 5))
msft0.plot(legend=True)
amd0.plot(legend=True)
```

```python
def plot_tickets(list_tickets, start_date = '2010-01-01'):
    for ticket in list_tickets:
        s = data.DataReader(ticket, 'google', start_date)['Close']
        s = (s - s.iloc[0]) / s.iloc[0]
        s.name = ticket
        s.plot(legend=True, figsize=(11, 5))
```

```python
plot_tickets(['aapl', 'msft', 'amd'])
```

```python
df_stocks = pd.concat([aapl, msft, amd], axis=1)
```

```python
df_stocks.isnull().sum()
```

```python
df_stocks.plot(figsize=(11, 5));
```

```python
df_stocks_pct = (df_stocks - df_stocks.iloc[0,:]) / df_stocks.iloc[0,:]
```

```python
df_stocks_pct.plot(figsize=(11,5), title='Stocks')
```
