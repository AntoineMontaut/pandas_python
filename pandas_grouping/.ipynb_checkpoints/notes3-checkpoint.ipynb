```python
import numpy as np
import pandas as pd
```

```python
coh = pd.read_csv('data/coh_employee.csv')
```

```python
pd.options.display.max_columns = 40
```

```python
coh.head(3)
```

# FILTER

```python
g = coh.groupby('POSITION_TITLE')
```

```python
def my_filter(df):
    return len(df) > 50
```

```python
g.filter(my_filter)['POSITION_TITLE'].unique()
```

```python
coh_filter = g.filter(my_filter)
```

```python
coh_filter.POSITION_TITLE.value_counts()
```

# TRANSFROM

```python
coh.pivot_table(index='RACE', columns='GENDER', values='BASE_SALARY', aggfunc='mean')
```

```python
coh_pivot = coh.pivot_table(index='RACE', columns='GENDER', values='BASE_SALARY', aggfunc=[np.mean, np.max, len])
coh_pivot
```

```python
coh.groupby('POSITION_TITLE').filter(lambda s: len(s)>50)\
    .pivot_table(index='RACE', columns=['POSITION_TITLE', 'GENDER'], values='BASE_SALARY', aggfunc=lambda x: round(np.mean(x),0))
```

```python
g = coh.groupby('POSITION_TITLE').filter(lambda df: len(df)>50)
```

```python
gg = g.groupby(['RACE', 'POSITION_TITLE', 'GENDER'])['BASE_SALARY'].mean()
gg = gg.unstack(['POSITION_TITLE', 'GENDER'])
gg.sort_index(axis=1, level='POSITION_TITLE')
gg
```
